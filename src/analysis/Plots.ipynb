```python
import pickle
weights = pickle.load(open('gitig_delme.pickle','rb'))
```

```python
%matplotlib inline

# https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def plot_sem_layer_weights(weights,title):

    fig, ax = plt.subplots()
    ax.tick_params(top=True, bottom=False,labeltop=True, labelbottom=False)
    im = ax.imshow(weights)
    cb = fig.colorbar(im)
    
    width = len(weights)

    plt.xticks(np.arange(0, width+1,10))
    plt.yticks(np.arange(0, width+1,10))

    ax.set_title(title)
    fig.tight_layout()
    plt.show()


# adam 0.0005, trained on sampling (1:1 pos:neg) + synthetic, train for 20 epochs
plot_sem_layer_weights(weights[0],'Trained after pretraining\n\n')
```
