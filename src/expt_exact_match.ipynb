```python
# import weights of semantic similarity layers
import pickle
separate_ablation = pickle.load(open('separate_ablation.pickle','rb'))
separate_full = pickle.load(open('separate_full.pickle','rb'))
shared_ablation = pickle.load(open('shared_ablation.pickle','rb'))
shared_full = pickle.load(open('shared_full.pickle','rb'))
```

```python
%matplotlib inline

# https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def plot_sem_layer_weights(weights,title):

    fig, ax = plt.subplots()
    ax.tick_params(top=True, bottom=False,labeltop=True, labelbottom=False)
    im = ax.imshow(weights)
    cb = fig.colorbar(im)
    
    width = len(weights)

    plt.xticks(np.arange(0, width+1,10))
    plt.yticks(np.arange(0, width+1,10))

    ax.set_title(title)
    fig.tight_layout()
    plt.show()


plot_sem_layer_weights(separate_ablation[0],'Separate encoder, no join and hidden layers\n')
plot_sem_layer_weights(separate_full[0],'Separate encoder, with join and hidden layers\n')
plot_sem_layer_weights(shared_ablation[0],'Shared encoder, no join and hidden layers\n')
plot_sem_layer_weights(shared_full[0],'Shared encoder, with join and hidden layers\n')
```
